```python
#It will be hard to extract data from the column above - on the first sight it looks that brand names are in the brackets and that the most
# useful information is contained anyway in the "Type" column
```

```python
# This is the code using the regex expressions that extracts brand names from the text inside the parentheses:
contr_other["brand name"] = contr_other["Type"].str.extract(r'\(([^)]+)\)')
```

```python
contr_other
```

```python
# this code didn't work for the cell, which contained more than 1 parentheses. 
# I will drop the "brand name" and "Explanation" column and try to find another solution to this problem
```

```python
contr_other = contr_other.drop(columns=["Explanation", "brand name"])
```

```python
import re

# Use findall to extract all text inside parentheses
contr_other["brand name"] = contr_other["Type"].apply(lambda x: re.findall(r'\(([^)]+)\)', x))
```
