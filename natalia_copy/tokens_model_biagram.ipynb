```python
review_for_processing = X_train[1]
review_for_processing
```

```python
from nltk.util import ngrams
```

```python
def preprocess_text_bi(text):
    tokeniser = RegexpTokenizer(r'\b\w+\b') #I keep only words and numbers - same what Celeste did
    tokens = tokeniser.tokenize(text)

    #Lower case and lemmatise
    lemmatiser = WordNetLemmatizer()
    lemmas = [lemmatiser.lemmatize(token.lower(), pos="v") for token in tokens]


    #Remove stop words
    keywords = [lemma for lemma in lemmas if lemma not in stopwords.words("english")]

    unigrams = keywords
    bigrams = [' '.join(bigram) for bigram in ngrams(keywords, 2)]
    
    return unigrams + bigrams
```

```python
tokens = preprocess_text_bi(text=review_for_processing) #now returning both bigrams and unigrams
```

```python
#preparing reviews for the model using tfidf
```

```python
vectoriser = TfidfVectorizer(analyzer=preprocess_text_bi) # this is the method you've implemented above
X_train_tfidf = vectoriser.fit_transform(X_train)  # notice that you are doing the fit only for train data
```

```python
X_test_tfidf = vectoriser.transform(X_test)
```
