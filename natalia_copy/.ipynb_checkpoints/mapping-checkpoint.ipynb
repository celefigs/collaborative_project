```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
test_df = pd.read_csv("./test_df.csv")
```

```python
reference_df = pd.read_csv("./reference_df.csv")
```

```python
reference_df = reference_df.drop(columns=["Unnamed: 0"])
```

```python
#function comparing drugName with both brand name and generic name

def find_match(drug_name, ref_df):
    
    for idx, row in ref_df.iterrows():
        brand_similarity = fuzz.ratio(drug_name, row['brand name'])
        generic_similarity = fuzz.ratio(drug_name, row['generic name'])
        
        if 70 <= brand_similarity <= 100 or 70 <= generic_similarity <= 100:
            return pd.Series([row['generic name'], row['type']])
    return pd.Series([None, None])

# Step 3: Apply the function to each row in test_df to find matching rows in reference_df
test_df[['matched_generic_name', 'matched_type']] = test_df['drugName'].apply(find_match, ref_df=reference_df)
```

```python
# Step 2: Create a function to compare drugName with both brand name and generic name
def find_match(drug_name, ref_df):
    # Check if drug_name is not NaN
    if pd.isna(drug_name):
        return pd.Series([None, None])

    for idx, row in ref_df.iterrows():
        # Ensure both brand name and generic name are not NaN
        brand_name = row['brand name'] if pd.notna(row['brand name']) else ''
        generic_name = row['generic name'] if pd.notna(row['generic name']) else ''
        
        # Compare drugName with brand name and generic name using fuzzy string matching
        brand_similarity = fuzz.ratio(drug_name, brand_name)
        generic_similarity = fuzz.ratio(drug_name, generic_name)
        
        # If the similarity is within the 70-80% range, return the generic name and type
        if 70 <= brand_similarity <= 80 or 70 <= generic_similarity <= 80:
            return pd.Series([row['generic name'], row['type']])
    
    return pd.Series([None, None])

# Step 3: Apply the function to each row in test_df to find matching rows in reference_df
test_df[['matched_generic_name', 'matched_type']] = test_df['drugName'].apply(find_match, ref_df=reference_df)
```
